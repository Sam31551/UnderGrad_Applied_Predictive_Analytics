Feature Engineering

```python
data.rename(columns={"def_wr": "defending_wr", "att_wr": "attacking_wr"})
```

```python
#Average attacking stats 
att_data = data.loc[:,['att' in i for i in data.columns]]
att_data.head(2)
```

```python
att_data.drop('att_wr', inplace=True, axis=1)
att_data.head(2)
```

```python
#Average defending stats
def_data = data.loc[:,['def' in i for i in data.columns]]
def_data.head(2)
```

```python
def_data.drop('def_wr', inplace=True, axis=1)
def_data.head(2)
```

```python
#Average attacking stats
att_data['average_attack'] = att_data.mean(axis=1)
att_data.head(5)
```

```python
att_data
```

```python
#Average defending stats
def_data['average_defend'] = def_data.mean(axis=1)
def_data.head(5)
```

```python
def_data
```

Add new columns to dataset

```python
avg_def = def_data['average_defend']
avg_def
```

```python
avg_att = att_data['average_attack']
avg_att
```

```python
data['Avg_Def'] = avg_def
data['Avg_Att'] = avg_att
data.head()
```

```python
data.columns
```

```python
#Remove ids, league_level, encoded vars
#All remaining variables re scored 1-99
continous_variables = ['overall', 'potential', 'value_eur', 'wage_eur', 'age', 'height_cm',
       'weight_kg', 'release_clause_eur', 'pace',
       'shooting', 'passing', 'dribbling', 'defending', 'physic',
       'attacking_crossing', 'attacking_finishing',
       'attacking_heading_accuracy', 'attacking_short_passing',
       'attacking_volleys', 'skill_dribbling', 'skill_curve',
       'skill_fk_accuracy', 'skill_long_passing', 'skill_ball_control',
       'movement_acceleration', 'movement_sprint_speed', 'movement_agility',
       'movement_reactions', 'movement_balance', 'power_shot_power',
       'power_jumping', 'power_stamina', 'power_strength', 'power_long_shots',
       'mentality_aggression', 'mentality_interceptions',
       'mentality_positioning', 'mentality_vision', 'mentality_penalties',
       'mentality_composure', 'defending_marking_awareness',
       'defending_standing_tackle', 'defending_sliding_tackle', 'Avg_Def', 'Avg_Att']
```

```python
unscalable_data = data[['club_team_id', 'league_level', 'club_position',
       'club_contract_valid_until', 'preferred_foot', 'weak_foot',
       'skill_moves', 'international_reputation', 'att_wr',
       'def_wr',]]
```

```python
scalable_data = data[['overall', 'potential', 'value_eur', 'wage_eur', 'age', 'height_cm',
       'weight_kg', 'release_clause_eur', 'pace',
       'shooting', 'passing', 'dribbling', 'defending', 'physic',
       'attacking_crossing', 'attacking_finishing',
       'attacking_heading_accuracy', 'attacking_short_passing',
       'attacking_volleys', 'skill_dribbling', 'skill_curve',
       'skill_fk_accuracy', 'skill_long_passing', 'skill_ball_control',
       'movement_acceleration', 'movement_sprint_speed', 'movement_agility',
       'movement_reactions', 'movement_balance', 'power_shot_power',
       'power_jumping', 'power_stamina', 'power_strength', 'power_long_shots',
       'mentality_aggression', 'mentality_interceptions',
       'mentality_positioning', 'mentality_vision', 'mentality_penalties',
       'mentality_composure', 'defending_marking_awareness',
       'defending_standing_tackle', 'defending_sliding_tackle', 'Avg_Def', 'Avg_Att']]
scalable_data.head()
```

Feature Scaling

```python
#Use MinMax Scaler as upper and lower bounds are known 
from sklearn.preprocessing import MinMaxScaler
minmaxsc = MinMaxScaler()

scaled_data = pd.DataFrame(minmaxsc.fit_transform(scalable_data.values), columns=scalable_data.columns, index=scalable_data.index)
scaled_data.head()
```

```python
# Place the DataFrames side by side
horizontal_stack = pd.concat([scaled_data, unscalable_data, ], axis=1)
horizontal_stack.head()
```
