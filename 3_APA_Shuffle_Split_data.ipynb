Dataset shuffling

```python
#Read in previously cleaned and scaled data
import pandas as pd
data = pd.read_csv('final_scaled_data.csv', index_col=False)
data
```

```python
data.drop(data.tail(1).index,inplace=True) 
```

```python
print(data.isnull().values.any())
```

Train Test Split

```python
from sklearn.model_selection import train_test_split

X=data.drop("value_eur", axis=1)
Y=data["value_eur"]
X
```

```python
Y
```

```python
#X_VAL = x_validation, same as x_test
X_train, X_val, y_train, y_val = train_test_split(X,
                                                  Y,
                                                  test_size=0.10,
                                                  shuffle=True,
                                                  random_state=324
                                                 )
```
